```python
#Importujemy odpowiednie biblioteki
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from pathlib import Path

np.random.seed(42) # Uczyńmy wyniki notatnika stablinymi w poszcególnych przebiegach

#Zdefinujmy domyślne rozmiary czcionek służące do generownaia rysunków
plt.rc('font', size=12)
plt.rc('axes', labelsize=14, titlesize=14)
plt.rc('legend', fontsize=12)
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10)

#zdefinujmy ustawienia bilbiloteki pandas
pd.set_option('display.max_columns', None) # wyświetlaj wszystkie kolumny w data
```

```python
datapath = Path() / "data" / "lifesat" #definujemy ścieżkę do bieżącego katalogu
#wczytujemy zbiory
oecd_bli = pd.read_csv(datapath / "oecd_bli.csv", sep=",")
gdp_per_capita = pd.read_csv(datapath / "gdp_per_capita.csv")
```

```python
#Wstępnie przetwarzamy dane PKB per capita tak, żeby wyłączyc dane z przed 2020 roku
gdp_per_capita_2020 = gdp_per_capita[gdp_per_capita["Year"]==2020]
gdp_per_capita_2020 = gdp_per_capita_2020.rename(
    columns={
        "GDP per capita, PPP (constant 2017 international $)"
        : "GDP per capita (USD)"}
)
gdp_per_capita_2020
```

```python
#poszukiwanie przez nas life satisfaction
oecd_bli["Indicator"].value_counts()
```

```python
#Tabela przestawna
oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]
oecd_bli = oecd_bli.pivot(
    index="Country",
    columns="Indicator",
    values="Value"
)
oecd_bli
```

```python
# Zresetujmy indeks, aby kraj był spowrotem kolumną. Przyda się to w późniejszych obliczeniach
oecd_bli = oecd_bli.reset_index()
oecd_bli
```

```python
#Wy swietlamy pdostawowe informajce o dataframe
gdp_per_capita_2020.info()
```

```python
oecd_bli.info()
```

```python
#ponieważ kolumny inaczej się nazywają ale są tego samego typu możemy zrobić ale tak wygląda merge
full_country_stats = oecd_bli.merge(gdp_per_capita_2020, left_on="Country", right_on="Entity", how="inner")
full_country_stats
```

```python
full_country_stats = full_country_stats[[
    "Country", "Life satisfaction", "GDP per capita (USD)"
]]
full_country_stats = full_country_stats.sort_values(by="GDP per capita (USD)")
full_country_stats
```

```python
full_country_stats.to_csv(datapath / "lifesat_full.csv", index=False)
```

```python
min_gdp = 23_500
max_gdp = 62_500
country_stats = full_country_stats[
    (full_country_stats["GDP per capita (USD)"] >= min_gdp)
    & (full_country_stats["GDP per capita (USD)"] <= max_gdp)
]
country_stats.to_csv(datapath / "lifesat.csv", index=False)
```

```python
#wczytujemy dane 
lifesat = pd.read_csv("data/lifesat/lifesat.csv")
lifesat.rename(columns={"GDP per capita (USD)":"PKB na mieszkańca"}, inplace=True)
lifesat.rename(columns={"Life satisfaction": "Satysfakcja z życia"}, inplace=True)
lifesat
```

```python
#WIZUALIZACJA DANYCH
lifesat.plot(kind='scatter', grid=True,
             x="PKB na mieszkańca", y="Satysfakcja z życia")
plt.show()
```

```python
#Przygotowujemy zmienne do modelu
X = lifesat[["PKB na mieszkańca"]]
y =lifesat[["Satysfakcja z życia"]]

#wybieramy model liniowy
model = LinearRegression()

#Trenujemy model 
model.fit(X,y)

#Obliczanie współczynniki modelu
theta_0=model.intercept_[0]
theta_1=model.coef_[0][0]
print(f"theta_0: {theta_0}")
print(f"theta_1: {theta_1}")
```

```python
lifesat.plot(kind='scatter', grid=True,
             x="PKB na mieszkańca", y="Satysfakcja z życia")
plt.plot(X, theta_0+theta_1*X)
plt.text(45000, 6.5, fr"$\theta_0 = {theta_0.round(2)}$",
         color = "blue")
plt.text(45000, 6.3, fr"$\theta_1 = {theta_1 * 1e5:.2f}\cdot 10^{{-5}}$",
         color = "blue")
plt.show()
```

```python
#Prognozy
cyprus_gdp_per_capita_2020 = (
    gdp_per_capita_2020[gdp_per_capita_2020["Entity"]=="Cyprus"][["GDP per capita (USD)"]]
)
cyprus_gdp_per_capita_2020.rename(columns={"GDP per capita (USD)":"PKB na mieszkańca"}, inplace=True)
cyprus_gdp_per_capita_2020
```

```python
cyprus_predicted_life_satisfactin = model.predict(cyprus_gdp_per_capita_2020)
cyprus_predicted_life_satisfactin
```

```python
#wybieramy model do regresji 3-najbliższych sąsiadów
from sklearn.neighbors import KNeighborsRegressor
model_knn = KNeighborsRegressor(n_neighbors=3)
#trenuemy model
model_knn.fit(X,y)
#Uzyskujemy przewidywania dla Cypru
model_knn.predict(cyprus_gdp_per_capita_2020)
```
